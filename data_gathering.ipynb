```python
import os
path = "./data/aggregated/transaction/country/india/state/"
agg_state_list = os.listdir(path)
agg_state_list
```

```python
import pandas as pd
import json

clm = {'id':[], 'state':[],'year':[],'quater':[],'transaction_type':[], 'transaction_count': [], 'transaction_amount':[]}
id = 1
for i in agg_state_list:
    p_i = path+i+"/"
    agg_yr = os.listdir(p_i)
    for j in agg_yr:
        p_j = p_i + j + "/"
        agg_yr_list = os.listdir(p_j)
        for k in agg_yr_list:
            p_k = p_j + k
            data = open(p_k, 'r')
            d = json.load(data)
            for z in d['data']['transactionData']:
                name = z['name']
                count = z['paymentInstruments'][0]['count']
                amount = z['paymentInstruments'][0]['amount']
                clm['transaction_type'].append(name)
                clm['transaction_count'].append(count)
                clm['transaction_amount'].append(amount)
                clm['state'].append(i)
                clm['year'].append(j)
                clm['quater'].append(int(k.strip('.json')))
                clm['id'].append(id)
                id = id +1

Agg_Trans = pd.DataFrame(clm)
```

```python
Agg_Trans
```

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+mysqlconnector://root:root@localhost:3306/phonepe_insights')
Agg_Trans.to_sql(name='aggregated_transactions_stats', con=engine, if_exists='replace', index=False)
engine.dispose()
```

## Aggregated User

```python
import os
import json
import pandas as pd

# agg_state_list should be a list of pulse-style state keys, e.g.:
# agg_state_list = ['maharashtra', 'andhra-pradesh', ...]
# base_dir should point to: .../data/aggregated/user/country/india/state
# final JSON paths: base_dir/<state>/<year>/<quarter>.json

base_dir = "./data/aggregated/user/country/india/state/"

rows = []
id_counter = 1

for state in agg_state_list:
    state_dir = os.path.join(base_dir, state)
    if not os.path.isdir(state_dir):
        continue

    for year in os.listdir(state_dir):
        year_dir = os.path.join(state_dir, year)
        if not os.path.isdir(year_dir):
            continue

        for qfile in os.listdir(year_dir):
            if not qfile.endswith(".json"):
                continue

            q = int(qfile.replace(".json", ""))   # quarter as int
            fpath = os.path.join(year_dir, qfile)

            try:
                with open(fpath, "r", encoding="utf-8") as f:
                    d = json.load(f)

                agg = d.get("data", {}).get("aggregated", {})
                # Guard against missing keys
                registered_users = int(agg.get("registeredUsers", 0) or 0)
                app_opens = int(agg.get("appOpens", 0) or 0)

                rows.append({
                    "id": id_counter,
                    "state": state,           # keep raw pulse key; pretty-map in UI
                    "year": int(year),
                    "quarter": q,
                    "total_users": registered_users,
                    "app_opens": app_opens
                })
                id_counter += 1

            except Exception as e:
                # Optional: collect bad paths for later inspection
                print(f"[WARN] Failed to parse {fpath}: {e}")

Agg_State = pd.DataFrame(rows, columns=["id","state","year","quarter","total_users","app_opens"])
```

```python
Agg_State
```

```python
from sqlalchemy import create_engine
engine = create_engine('mysql+mysqlconnector://root:root@localhost:3306/phonepe_insights')
Agg_State.to_sql(name='aggregated_user_state_totals', con=engine, if_exists='replace', index=False)
```
